## Data Exploration & Plot

```python
# Import Libraries.
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load dataset from URL.
df = pd.read_csv("https://data.cityofchicago.org/api/views/9xs2-f89t/rows.csv")

# Basic Exploration: Shape & preview.
print("Dataset Shape:", df.shape)  # (566, 79).
```

```python
df.head()
```

```python
# Check data types.
df.dtypes  # Mix of object (Strings), int64, float64.
```

```python
# Summary statistics for numeric columns.
df.describe()
```

```python
# Exploration: Average attendance by school type.
attendance_by_type = df.groupby("Elementary, Middle, or High School")["Average Student Attendance"].mean()
print("Average Student Attendance by School Type:\n", attendance_by_type)
```

```python
# Clean data for plotting (Drop rows with missing attendance or test scores).
df_clean = df.dropna(subset = ["Average Student Attendance", "ISAT Exceeding Reading % "])

# Plot: Attendance vs. Reading Scores.
plt.scatter(df_clean["Average Student Attendance"], df_clean["ISAT Exceeding Reading % "], alpha = 0.5)
plt.xlabel("Average Student Attendance (%)")
plt.ylabel("ISAT Reading Exceeding (%)")
plt.title("Attendance vs. Reading Performance (2011 - 2012)")
plt.show();
```

```python
# Additional Exploration: Top 5 safest schools.
safety_scores = df[["Name of School", "Safety Score"]].sort_values(by = "Safety Score", ascending = False)
print("Top 5 Schools by Safety Score:\n", safety_scores.head())
```
